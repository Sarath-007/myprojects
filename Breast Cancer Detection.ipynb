```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
df_cancer=pd.DataFrame(np.c_[cancer['data'],cancer['target']],columns=np.append(cancer['feature_names'],['target']))
```

```python
df_cancer.head()
```

```python
df_cancer.shape
```

```python
df_cancer.tail()
```

```python
df_cancer.describe()
```

```python
df_cancer.info()
```

```python
# model training
x= df_cancer.drop(['target'],axis=1)
```

```python
y=df_cancer['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=5)
```

```python
x_train
```

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix , classification_report , accuracy_score
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
```

```python
y_predict
```

```python
cm=confusion_matrix(y_test,y_predict)
sns.heatmap(cm, annot=True)
```

```python
accuracy_score(y_test,y_predict)
```

```python
y_test.shape
```

```python
print(classification_report(y_test,y_predict))
```

# USING SEQUENTIAL MODEL

```python
import keras
from keras.utils import to_categorical
y=to_categorical(y)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
s_model=Sequential()

s_model.add(Dense(128,activation="relu",input_shape=(30,)))
s_model.add(Dense(64,activation="relu"))
s_model.add(Dense(2,activation="softmax"))
```

```python
s_model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
s_model.fit(x,y,epochs=50)
```
